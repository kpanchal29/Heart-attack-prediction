```python
# Replace "?" with NaN
df.replace("?", np.nan, inplace=True)

# Convert numeric columns stored as object to proper numeric types
cols_to_convert = ['trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'slope', 'ca', 'thal']
df[cols_to_convert] = df[cols_to_convert].apply(pd.to_numeric, errors='coerce')

# Check how many missing values we have after conversion
missing_values = df.isnull().sum()
missing_values
```

```python
# Drop columns with excessive missing data
df_cleaned = df.drop(columns=['slope', 'ca', 'thal'])

# Drop rows with any remaining missing values
df_cleaned = df_cleaned.dropna()

# Confirm the dataset is now clean
df_cleaned.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Clean column names
df_cleaned.columns = df_cleaned.columns.str.strip()
```

```python
plt.figure(figsize=(8, 5))
df_cleaned['num'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Heart Disease Severity Distribution')
plt.xlabel('Heart Disease Severity (0 = No Disease)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
labels = ['Female', 'Male']
sizes = df_cleaned['sex'].value_counts()
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='num', y='chol', data=df_cleaned)
plt.title('Cholesterol Levels by Heart Disease Severity')
plt.xlabel('Heart Disease Severity')
plt.ylabel('Cholesterol')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x='sex', y='thalach', data=df_cleaned)
plt.title('Max Heart Rate Distribution by Sex')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Maximum Heart Rate')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.swarmplot(x='cp', y='age', hue='num', data=df_cleaned)
plt.title('Age and Chest Pain Type by Heart Disease')
plt.xlabel('Chest Pain Type')
plt.ylabel('Age')
plt.legend(title='Heart Disease')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='chol', y='thalach', hue='num', data=df_cleaned)
plt.title('Cholesterol vs Max Heart Rate (colored by Heart Disease)')
plt.xlabel('Cholesterol')
plt.ylabel('Max Heart Rate')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
corr = df_cleaned.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.stripplot(x='num', y='oldpeak', data=df_cleaned, jitter=True)
plt.title('ST Depression (oldpeak) by Heart Disease')
plt.xlabel('Heart Disease Severity')
plt.ylabel('Oldpeak')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df_cleaned, x='age', hue='num', fill=True)
plt.title('Age Distribution by Heart Disease (KDE)')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
df_cleaned['age_group'] = pd.cut(df_cleaned['age'], bins=[25, 35, 45, 55, 65, 75], labels=['25-34','35-44','45-54','55-64','65-74'])
avg_hr = df_cleaned.groupby('age_group')['thalach'].mean().reset_index()
sns.lineplot(data=avg_hr, x='age_group', y='thalach', marker='o')
plt.title('Average Max Heart Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Max Heart Rate')
plt.show()
```
